## Análisis de valores atípicos

Se identifica valores atípicos en las variables numéricas utilizando métodos visuales (diagramas de caja) y/o estadísticos, y se discute posibles estrategias para su tratamiento.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Seleccionar columnas numéricas relevantes (excluyendo 'group_id')
numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
numeric_cols.remove('group_id')

print("Generando diagramas de caja para las columnas numéricas:")

# Iterar sobre las columnas numéricas y crear diagramas de caja
for col in numeric_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Diagrama de Caja de {col}')
    plt.xlabel(col)
    plt.show()
```

Discusión sobre estrategias de tratamiento de valores atípicos:
- participation: El diagrama de caja muestra algunos valores atípicos en el extremo superior. Dado que representa un porcentaje, valores superiores al 100% son claramente erróneos y deberían ser investigados o eliminados. Otros valores altos podrían ser legítimos pero extremos. Posibles estrategias: investigación de valores > 100%, winsorización o transformación si la distribución es muy sesgada.
- estimado_ingresos: El diagrama de caja muestra valores atípicos en el extremo superior. Estos podrían representar transacciones de alto valor. Posibles estrategias: winsorización, transformación logarítmica para reducir el impacto de valores extremos, o mantener si son legítimos y el modelo a utilizar es robusto a outliers.
- limite_max_res: Similar a 'estimado_ingresos', presenta valores atípicos superiores. Podrían ser límites legítimamente altos. Posibles estrategias: winsorización o transformación logarítmica.
- prima_neta: También muestra valores atípicos superiores, incluyendo algunos en cero que podrían requerir investigación. Posibles estrategias: winsorización, transformación logarítmica (manejar valores cero si se aplica), o mantener si son legítimos.

## Análisis de cardinalidad


Se evalua la cardinalidad de las variables categóricas y considera estrategias para manejar aquellas con alta cardinalidad.


```python
# Seleccionar columnas de tipo 'object' o 'category'
categorical_cols = df.select_dtypes(include=['object', 'category']).columns

print("Evaluando la cardinalidad de las variables categóricas:\n")

# Calcular y mostrar el número de valores únicos para cada columna categórica
cardinality = {}
high_cardinality_cols = []
for col in categorical_cols:
    num_unique = df[col].nunique()
    cardinality[col] = num_unique
    print(f"Columna '{col}': {num_unique} valores únicos")
    if num_unique > 50: # Umbral arbitrario para alta cardinalidad
        high_cardinality_cols.append(col)

print("\nColumnas con alta cardinalidad (más de 50 valores únicos):")
if high_cardinality_cols:
    for col in high_cardinality_cols:
        print(f"- {col}: {cardinality[col]} valores únicos")
        # Mostrar una muestra de valores únicos para columnas con alta cardinalidad
        print("  Muestra de valores únicos:")
        display(df[col].unique()[:10]) # Mostrar los primeros 10 valores únicos
else:
    print("No se encontraron columnas con alta cardinalidad (más de 50 valores únicos).")
```

Discusión de estrategias para manejar variables categóricas con alta cardinalidad:
Las columnas con alta cardinalidad identificadas son 'reinsurer', 'asegurado', 'direccion', 'actividad', 'interes_asegurable', 'riesgos_cubiertos', 'condiciones_reaseguro', 'exclusiones', 'soporte', 'anexos_clausulas', y 'panel_participantes'.
Para manejar estas columnas, se pueden considerar las siguientes estrategias:
- **Codificación por Frecuencia:** Reemplazar cada categoría con su frecuencia de aparición. Útil cuando la relación entre la categoría y la variable objetivo se relaciona con su frecuencia.
- **Agrupación de Categorías:** Agrupar categorías poco frecuentes en una sola categoría 'Otros'. Requiere conocimiento del dominio o análisis adicional.
- **Codificación de Hash:** Mapear categorías a un espacio de menor dimensión usando una función hash. Puede haber colisiones, pero es útil para un gran número de categorías.
- **Eliminación:** Si la columna no se considera relevante para el análisis o modelado, o si las otras estrategias no son viables.
- **Codificación basada en Target:** Para tareas de aprendizaje supervisado, codificar categorías basado en la media de la variable objetivo para cada categoría. Requiere precaución para evitar el sobreajuste.

La elección de la estrategia dependerá del contexto del análisis y los modelos que se planeen utilizar.

## Análisis univariante

Se visualiza la distribución de variables individuales (numéricas y categóricas) utilizando histogramas, diagramas de barras, etc.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 1. Visualización de variables numéricas (Histogramas)
numerical_cols = ['participation', 'estimado_ingresos', 'limite_max_res', 'prima_neta']

# Manejar valores faltantes en 'participation' para la visualización
# Usaremos la mediana para imputar temporalmente para el histograma
participation_median = df['participation'].median()
df_participation_filled = df['participation'].fillna(participation_median)

print("Generando histogramas para variables numéricas:")

for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    if col == 'participation':
        sns.histplot(df_participation_filled, kde=True)
        plt.title(f'Distribución de {col} (Valores Faltantes Imputados con Mediana)')
    else:
        sns.histplot(df[col], kde=True)
        plt.title(f'Distribución de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.show()

```

Interpretación de los histogramas:

- limite_max_res: Similar a estimado_ingresos, esta distribución también muestra un sesgo pronunciado hacia la derecha.
- prima_neta: La distribución de la prima neta también está sesgada a la derecha, indicando que la mayoría de las primas son de valores más bajos.


```python
# 2. Visualización de variables categóricas (Diagramas de barras)
categorical_cols_to_plot = ['record_type', 'tipo', 'territorialidad', 'jurisdiccion', 'terminos_pago']

print("\nGenerando diagramas de barras para variables categóricas:")

for col in categorical_cols_to_plot:
    plt.figure(figsize=(12, 7)) # Ajustar tamaño para legibilidad
    sns.countplot(data=df, y=col, order=df[col].value_counts().index, palette='viridis')
    plt.title(f'Frecuencia de Categorías en {col}')
    plt.xlabel('Frecuencia')
    plt.ylabel(col)
    plt.tight_layout() # Ajustar para evitar superposición
    plt.show()
```

Interpretación de los diagramas de barras:

```python
# 4. Visualizar tendencias temporales de variables numéricas (por año)
print("\nVisualizando tendencias temporales de variables numéricas por año de inicio:")

# Variables numéricas de interés
numeric_vars_time = ['estimado_ingresos', 'prima_neta']

# Agrupar por año de inicio y calcular la suma y el promedio
df_grouped_year = df.groupby(df['periodo_inicio'].dt.year)[numeric_vars_time].agg(['sum', 'mean'])

# Renombrar las columnas para mayor claridad en los gráficos
df_grouped_year.columns = ['_'.join(col).strip() for col in df_grouped_year.columns.values]

# Trazar gráficos de línea para la suma y el promedio
for var in numeric_vars_time:
    plt.figure(figsize=(14, 7))

    # Gráfico de suma
    plt.subplot(1, 2, 1) # 1 fila, 2 columnas, primer gráfico
    sns.lineplot(data=df_grouped_year, x=df_grouped_year.index, y=f'{var}_sum', marker='o')
    plt.title(f'Suma Total de {var} por Año de Inicio')
    plt.xlabel('Año de Inicio')
    plt.ylabel(f'Suma Total de {var}')
    plt.grid(True, linestyle='--', alpha=0.6)

    # Gráfico de promedio
    plt.subplot(1, 2, 2) # 1 fila, 2 columnas, segundo gráfico
    sns.lineplot(data=df_grouped_year, x=df_grouped_year.index, y=f'{var}_mean', marker='o', color='orange')
    plt.title(f'Promedio de {var} por Año de Inicio')
    plt.xlabel('Año de Inicio')
    plt.ylabel(f'Promedio de {var}')
    plt.grid(True, linestyle='--', alpha=0.6)

    plt.tight_layout() # Se ajusta para evitar superposición
    plt.show()
```

Interpretación de las tendencias temporales de variables numéricas:
- Los gráficos de línea muestran la evolución de la suma total y el promedio de 'estimado_ingresos' y 'prima_neta' a lo largo de los años de inicio.
- Observar si hay tendencias crecientes, decrecientes o estacionales en estas métricas.
- Comparar las tendencias de la suma total y el promedio puede indicar si los cambios en el volumen total se deben a un aumento en el número de reaseguros o a un cambio en el valor promedio por reaseguro.
- Por ejemplo, si la suma total aumenta pero el promedio se mantiene estable, sugiere un aumento en el número de reaseguros. Si el promedio aumenta, sugiere que los reaseguros individuales son de mayor valor.
